Data can be found [here](https://www.kaggle.com/dansbecker/melbourne-housing-snapshot)

```python
import pandas as pd

data = pd.read_csv('melb_data.csv')

columns_to_use = ['Rooms','Distance','Landsize','BuildingArea','YearBuilt']
X = data[columns_to_use]
y = data['Price']
```

Notes_210922_110800.jpg

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

my_pipeline = Pipeline(steps=[('preprocessor',SimpleImputer()),
                              ('model',RandomForestRegressor(n_estimators=50, random_state=0))]) 
```

We use **cros_val_score()** fuction to measure the performance in each experiment.

```python
from sklearn.model_selection import cross_val_score
# multiplied by (-1) since this function calculates negative MAE
scores = -1*cross_val_score(my_pipeline,X,y,cv=3,scoring='neg_mean_absolute_error')
print(f" here are the scores for each experimnet: {scores}")
print(f" and here is the average: {scores.mean()} ")
```
