```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
x = iris.data[:, 2:]
y = iris.target

tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(x,y)
```

**Visualizing Decision Tree**

```python
#visualizing the tree
from sklearn import tree
plt.figure(figsize=(10,8))
tree.plot_tree(tree_clf,filled=True,feature_names=iris.feature_names[2:],
               class_names=iris.target_names,rounded=True,proportion=True,node_ids=True,)
```

```python
print(tree.export_text(tree_clf,feature_names=iris.feature_names[2:]))
```

```python
print(tree_clf.predict_proba([[1,1]]))
tree_clf.predict([[1,1]])
```
